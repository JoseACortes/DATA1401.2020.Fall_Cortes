Write a function that takes in a list of files with dates as filenames and return a list with the filenames sorted by date.

```python
import copy
all_files_reversed = all_files.copy()
all_files_reversed.reverse()
```

```python
all_files_reversed
```

```python
def datesort(filelist):
    sorte = list()
    for file in filelist:
        filedate = file.split("/")[-1].split(".")[0].split("\\")[1]
        templist = sorte
        dateval = time.mktime(datetime.datetime.strptime(filedate,"%m-%d-%Y").timetuple())
        last = True
        for n, s in enumerate(sorte):
            s = time.mktime(datetime.datetime.strptime(s.split("/")[-1].split(".")[0].split("\\")[1],"%m-%d-%Y").timetuple())
            if dateval < s:
                last = False
                templist[n+1:len(sorte)+1] = templist[n:len(sorte)]
                templist[n] = file
                break
        if last == True:
            templist.append(file)
        sorte = templist
    return sorte
```

Write a function that takes the sorted list of files from previous exercise, and outputs a data structure analogous to this example, containing all of the data stored in the files.

```python
def getlocationlist(frame):
    locations = list()
    try:
        for n, d in enumerate(frame['Combined_Key']):
            locations.append(str(d))
    except:
        for n, d in enumerate(frame['Province/State']):
            if str(d) == 'nan':
                locations.append(str(frame['Country/Region'][n]))
            else:
                locations.append(d+', '+str(frame['Country/Region'][n]))
    return locations

def addemptyall(currdict = {}):
    for word in currdict:
        currdict[word]['Confirmed'].append(0)
        currdict[word]['Deaths'].append(0)
        

def checkandaddword(currdict = {}, newword = 'no name error', currentlength = 0):
    if newword in currdict:
        pass
    else:
        currdict[newword] = {'Confirmed': [0]*(currentlength+1), 'Deaths': [0]*(currentlength+1)}
            
```

```python
def threedimensionalrestructure(filelist):
    newstruct = dict()
    for n, f in enumerate(filelist):
        addemptyall(newstruct)
        currframe = pd.read_csv(f)
        locationlist = getlocationlist(currframe)
        confirmlist = currframe['Confirmed']
        deathlist = currframe['Deaths']
        for x, location in enumerate(locationlist):
            checkandaddword(newstruct, location, n)
            if str(confirmlist[x]) == 'nan':
                newstruct[location]['Confirmed'][-1] = 0
            else:
                newstruct[location]['Confirmed'][-1] = int(confirmlist[x])
            if str(deathlist[x]) == 'nan':
                newstruct[location]['Deaths'][-1] = 0
            else:
                newstruct[location]['Deaths'][-1] = int(deathlist[x])
            #newstruct[location]['Confirmed'][-1] = confirmlist[x]
            #newstruct[location]['Deaths'][-1] = deathlist[x]
    return newstruct
```

**Question 5**: Write a function that takes a list of region names and the data structure from previous question and aggregate the fields over the regions, outputing a data structure that is otherwise identical to the input. Use this function to aggregate the data for each country.

```python
def feildsum(struct, reigons):
    newstruct = {}
    datalen = len(struct[next(iter(struct))]['Confirmed'])
    for r in reigons:
        currreigonConfirmed = [0]*datalen
        currreigonDeaths = [0]*datalen
        for n, name in enumerate(struct):
            if r == str(name).split(', ')[-1]:
                newcon = list()
                newdead = list()
                for x, c in enumerate(struct[name]['Confirmed']):
                    newcon.append(currreigonConfirmed[x]+c)
                    newdead.append(currreigonDeaths[x]+struct[name]['Deaths'][x])
                currreigonConfirmed = newcon
                currreigonDeaths = newdead
        newstruct[r] = {'Confirmed': currreigonConfirmed, 'Deaths': currreigonDeaths}
    return newstruct
```

**Question 6**: Develop an algorithm that identifies when COVID infections started in each region. Create a time ordered list of regions and countries of when COVID started.


```python
def startcheck(reigonframe):
    ordered = list()
    orderedvals = list()
    for r in reigonframe:
        t = 0
        for n, c in enumerate(reigonframe[r]['Confirmed']):
            if c > 0:
                t = n
                break
        otemp = orderedvals
        k = False
        for x, o in enumerate(otemp):
            if t < o:
                k = True
                orderedvals[x+1:] = orderedvals[x:]
                orderedvals[x] = t
                ordered[x+1:] = ordered[x:]
                ordered[x] = r
                break
        if k == False:
            orderedvals.append(t)
            ordered.append(r)
    return ordered
```

```python
startcheck(reigonstotal)
```

**Question 7**: Develop an algorithm that identifies peaks in COVID infections. Use the algorithm to create lists of regions ordered by when peaks occured. Create ordered lists for regions for the first and second wave.

```python
def detectpeaks(chlist):
    firstpeak = 0
    seccondpeak = 0
    firstpeakfound = False
    seccondpeakfound = False
    for x, i in enumerate(chlist[:-2]):
        if i > chlist[x+1]:
            firstpeak = x
            firstpeakfound = True
            for y, v in enumerate(chlist[x+1:-2]):
                if v>chlist[y+1]:
                    seccondpeak = y
                    seccondpeakfound = True
                    break
            break
    if firstpeakfound == True and seccondpeak == False:
        seccondpeak = len(chlist)
    elif firstpeakfound == False:
        firstpeak = len(chlist)
        seccondpeak = firstpeak
    return(firstpeak, seccondpeak)
```

```python
def peakdetection(reigonframe):
    firstpeakorderedlist = list()
    firstpeakvallist = list()
    seccondpeakorderedlist = list()
    seccondpeakvallist = list()
    for r in reigonframe:
        currentreigonvals = detectpeaks(reigonframe[r]['Confirmed'])
        avallist = firstpeakvallist
        bvallist = seccondpeakvallist
        afound = False
        bfound = False
        for x, a in enumerate(avallist):
            if currentreigonvals[0]<a:
                afound = True
                firstpeakvallist[x+1:] = firstpeakvallist[x:]
                firstpeakvallist[x] = currentreigonvals[0]
                firstpeakorderedlist[x+1:] = firstpeakorderedlist[x:]
                firstpeakorderedlist[x] = r
                break
        if afound == False:
            firstpeakvallist.append(currentreigonvals[0])
            firstpeakorderedlist.append(r)
        for y, b in enumerate(bvallist):
            if currentreigonvals[1]<b:
                bfound = True
                seccondpeakvallist[y+1:] = seccondpeakvallist[y:]
                seccondpeakvallist[y] = currentreigonvals[1]
                seccondpeakorderedlist[y+1:] = seccondpeakorderedlist[y:]
                seccondpeakorderedlist[y] = r
                break
        if bfound == False:
            seccondpeakvallist.append(currentreigonvals[0])
            seccondpeakorderedlist.append(r)
    return(firstpeakorderedlist, seccondpeakorderedlist)
```
